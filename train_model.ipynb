```python
from datamodule import CatDogDataModule
from model import ImagePredictionLogger, BaseClassifier
from pytorch_lightning.loggers import WandbLogger
import pytorch_lightning as pl
```

```python
data = CatDogDataModule(root='data', batch_size=32)
model = BaseClassifier(lr=0.01)
samples = next(iter(data.val_dataloader()))
```

```python
wandb_logger = WandbLogger(project="lit-wandb")
```

```python
trainer = pl.Trainer(
    accelerator='mps',
    devices=1,
    logger=wandb_logger,    # W&B integration
    log_every_n_steps=5,   # set the logging frequency
    val_check_interval=10,
    max_steps=100,
    deterministic=True,     # keep it deterministic
    callbacks=[ImagePredictionLogger(samples)] # see Callbacks section
    )
```

```python
trainer.fit(model, data)
```
